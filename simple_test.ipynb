```python
from src.simulate import simulate_data
from src.threshold_clustering import threshold_clustering
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = simulate_data(3, 200, 3, 2)
colors = ['red', 'green', 'blue']
assigned_colors = {l:color for l, color in zip(np.unique(data['labels']), colors)}
labels_colors = [assigned_colors[l] for l in data['labels']]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(data[0], data[1], data[2], color = labels_colors)
```

```python
predicted_labels = threshold_clustering(data.loc[:, data.columns != 'labels'], 40, 3)
predicted_labels
```

```python
predicted_labels_colors = [assigned_colors[l] for l in predicted_labels]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(data[0], data[1], data[2], color = predicted_labels_colors)
```
